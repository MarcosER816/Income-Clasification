```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sklearn
```

```python
from pandas import read_csv
```

```python
path=("/content/us-adult-income-update/census.csv")
data=read_csv(path)
data.head()
```

```python
df=pd.DataFrame(data)
df
```

```python
print(f'Number of Rows: {df.shape[0]}')
print(f'Number of Columns: {df.shape[1]}')
```

```python
for i, col in enumerate(df.columns):
  print(df.columns[i], '::\n', df[str(col)].unique(), '\n\n')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.isin(['?']).sum(axis=0)
```

```python
df.replace("?", np.nan, inplace = True)
df.isna().sum()
```

```python
df=df.dropna()
df.isna().sum()
```

```python
df
```

```python
from IPython.display import display

for columna in df.columns:

    tabla = (
        df[columna]
        .value_counts(dropna=False)
        .reset_index()
    )

    tabla.columns = [columna, "Cantidad"]

    print(f"\n\n===== {columna} =====")
    display(tabla)
```

```python
df["income"] = df["income"].map({
    "<=50K": 0,
    ">50K": 1
})
```

```python
y = df["income"]
x = df.drop("income", axis=1)
x.head()
```

```python
# Columnas categóricas
columnas_categoricas = x.select_dtypes(include=['object']).columns

print(columnas_categoricas)
```

```python
x = pd.get_dummies(
    x,
    columns=columnas_categoricas,
    dtype=int
)

x.head()
```

```python
df_corr = x.copy()

df_corr["income"] = y

correlaciones = df_corr.corr()["income"]

correlaciones = correlaciones.abs().sort_values(
    ascending=False
)

print(correlaciones.head(50))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_income = (
    df_corr
    .corr()[["income"]]
    .sort_values("income", ascending=False)
)

plt.figure(figsize=(8,25))

sns.heatmap(
    corr_income,
    annot=True,
    cmap="coolwarm",
    center=0
)

plt.title("Correlación de cada variable con income")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x,
    y,
    train_size=0.8,
    random_state=2287
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

modelo = Pipeline([

    (
        "scaler",
        StandardScaler()
    ),

    (
        "model",
        LogisticRegression(
            solver='lbfgs',
            penalty='l2',
            C=1,
            max_iter=5000
        )
    )

])
```

```python
modelo.fit(X_train, y_train)
```

```python
print(f"Training accuracy: {modelo.score(X_train,y_train):.4f}")

print(f"Test accuracy: {modelo.score(X_test,y_test):.4f}")
```

```python
y_pred_train = modelo.predict(X_train)

y_pred_test = modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc_train = accuracy_score(
    y_train,
    y_pred_train
)

acc_test = accuracy_score(
    y_test,
    y_pred_test
)

print("Accuracy entrenamiento:", acc_train)

print("Accuracy prueba:", acc_test)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

precision = precision_score(
    y_test,
    y_pred_test
)

recall = recall_score(
    y_test,
    y_pred_test
)

f1 = f1_score(
    y_test,
    y_pred_test
)

print("Precision:", precision)
print("Recall:", recall)
print("F1:", f1)
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        y_pred_test
    )
)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_test,
    y_pred_test
)

print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicción")

plt.ylabel("Valor Real")

plt.title("Matriz de Confusión")

plt.show()
```

```python
from sklearn.metrics import roc_auc_score

y_pred_probs = modelo.predict_proba(X_test)

roc_auc = roc_auc_score(
    y_test,
    y_pred_probs[:,1]
)

print("ROC-AUC:", roc_auc)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(
    y_test,
    y_pred_probs[:,1]
)

plt.figure(figsize=(7,5))

plt.plot(
    fpr,
    tpr,
    linewidth=2
)

plt.plot(
    [0,1],
    [0,1],
    '--'
)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curva ROC")

plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score
)

parametros_rf = {

    'n_estimators':[100,200,300],

    'max_depth':[10,20,None],

    'min_samples_split':[2,5,10],

    'min_samples_leaf':[1,2,4]

}

rf_base = RandomForestClassifier(

    random_state=2287,

    n_jobs=-1

)

grid_rf = GridSearchCV(

    estimator=rf_base,

    param_grid=parametros_rf,

    cv=5,

    scoring='f1',

    n_jobs=-1,

    verbose=2

)

grid_rf.fit(
    X_train,
    y_train
)

print("Mejores parámetros RF:")
print(grid_rf.best_params_)

rf = grid_rf.best_estimator_

y_pred_rf = rf.predict(
    X_test
)

print(
    classification_report(
        y_test,
        y_pred_rf
    )
)
```

```python
pd.DataFrame(
    grid_rf.cv_results_
)[
    [
        'mean_test_score',
        'params'
    ]
].sort_values(
    'mean_test_score',
    ascending=False
).head(10)
```

```python
from sklearn.metrics import accuracy_score

print(
    "Accuracy RF:",
    accuracy_score(
        y_test,
        y_pred_rf
    )
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm_rf = confusion_matrix(
    y_test,
    y_pred_rf
)

plt.figure(figsize=(6,6))

sns.heatmap(
    cm_rf,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.title(
    "Random Forest"
)

plt.xlabel("Predicción")
plt.ylabel("Valor Real")

plt.show()
```

```python
importancias = pd.DataFrame({

    "Variable": X_train.columns,

    "Importancia": rf.feature_importances_

})

importancias = importancias.sort_values(

    by="Importancia",

    ascending=False

)

importancias.head(20)
```

```python
top = importancias.head(15)

plt.figure(figsize=(10,6))

sns.barplot(

    data=top,

    x="Importancia",

    y="Variable"

)

plt.title(
    "Top 15 Variables Más Importantes"
)

plt.show()
```

# **Aumentando clase minoritaria**

```python
print(y_train.value_counts())
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(
    sampling_strategy='auto',
    random_state=2287
)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

print("Distribución original:")
print(y_train.value_counts())

print("\nDistribución después de SMOTE:")
print(y_train_smote.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_smote_scaled = scaler.fit_transform(
    X_train_smote
)

X_test_scaled = scaler.transform(
    X_test
)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

parametros_svm = {

    'C':[0.1,1,2,5],

    'gamma':[
        'scale',
        0.1,
        0.01
    ],

    'kernel':[
        'linear',
        'rbf'
    ]

}

svm_base = SVC(

    probability=True,

    random_state=2287

)

grid_svm = GridSearchCV(

    estimator=svm_base,

    param_grid=parametros_svm,

    cv=5,

    scoring='f1',

    n_jobs=-1,

    verbose=2

)

grid_svm.fit(

    X_train_smote_scaled,

    y_train_smote

)

print("Mejores parámetros SVM:")
print(grid_svm.best_params_)

svm_model = grid_svm.best_estimator_
```

```python
pd.DataFrame(
    grid_svm.cv_results_
)[
    [
        'mean_test_score',
        'params'
    ]
].sort_values(
    'mean_test_score',
    ascending=False
).head(10)
```

```python
print(y_train_smote.value_counts())
```

# **Red neuronal (NN)**

```python
from sklearn.preprocessing import StandardScaler

scaler_nn = StandardScaler()

X_train_scaled = scaler_nn.fit_transform(
    X_train
)

X_test_scaled = scaler_nn.transform(
    X_test
)
```

```python
from sklearn.neural_network import MLPClassifier

nn = MLPClassifier(

    hidden_layer_sizes=(128,64,32),

    activation='relu',

    solver='adam',

    alpha=0.0001,

    batch_size=128,

    learning_rate='adaptive',

    max_iter=500,

    random_state=2287

)
```

```python
nn.fit(
    X_train_scaled,
    y_train
)
```

# **resultados NN**

```python
y_pred_nn = nn.predict(
    X_test_scaled
)

print(
    classification_report(
        y_test,
        y_pred_nn
    )
)
```

```python
from sklearn.metrics import accuracy_score

print(
    "Accuracy NN:",
    accuracy_score(
        y_test,
        y_pred_nn
    )
)
```

```python
from sklearn.metrics import roc_auc_score

y_prob_nn = nn.predict_proba(
    X_test_scaled
)

roc_nn = roc_auc_score(
    y_test,
    y_prob_nn[:,1]
)

print(
    "ROC AUC NN:",
    roc_nn
)
```
